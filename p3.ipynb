## **Questão 1**: Polinômios interpoladores
Nesta questão, iremos trabalhar com o espaço de funções reais $\mathcal{F}$, e também com o espaço dos polinômios de grau até $d$, denotado por $\mathcal{P}_d$.

### Item 1.a) 
O **funcional de avaliação**. Seja $X = \set{x_1,...,x_m}$ um conjunto de pontos distintos em $\mathbb{R}$. Defina o funcional de avaliação $L_X : \mathcal{F} → \mathbb{R}^m$ por $L_X (f) = (f(x_1),...,f(x_m))$. Mostre que $L_X$ é uma transformação linear.

Para mostrar que $L_X$ é uma transformação linear basta mostrar que $L_X(\alpha f + \beta g) = \alpha L_X(f) + \beta L_X(g), \forall \alpha, \beta \in \mathbb{R}, f, g \in \mathcal{F}$ 

$$ 
\begin{aligned}
L_X(\alpha f + \beta g) &= ((\alpha f + \beta g)(x_1),..., (\alpha f + \beta g)(x_m)) \\
&= (\alpha f(x_1),...,\alpha f(x_m)) + (\beta g(x_1),...,\beta g(x_m)) \\
&= \alpha (f(x_1),...,f(x_m)) + \beta (g(x_1),...,g(x_m)) \\
&= \alpha L_X(f) + \beta L_X(g)
\end{aligned}
$$

### Item 1.b) 
Podemos ver o espaço dos polinômios $\mathcal{P}_d$ como um subconjunto de $\mathcal{F}$. Mostre que de fato é um subespaço linear, e conclua que $L_X$ restrita a $\mathcal{P}_d$ é uma transformação linear. (Dica: não é necessário usar uma base de $\mathcal{P}_d$!)

Parte 01: Mostrar que $\mathcal{P}_d$ é um subespaço linear

Sejam $p, q \in \mathcal{P}_d$, logo ambos tem grau menor ou igual à $d$ e, para $a_i, b_i \in \mathbb{R}$, podemos escreve-los da seguinte forma:
$$
p(x) = \sum^d_{i=0}{a_i x^i}
\\
q(x) = \sum^d_{i=0}{b_i x^i}
$$

Agora, dados $\alpha, \beta \in \mathbb{R}$, temos que:
$$
\begin{aligned}
\alpha p(x) + \beta q(x) &= \alpha \sum^d_{i=0}{a_i x^i} + \beta \sum^d_{i=0}{b_i x^i} 
\\
&= \sum^d_{i=0}{\alpha a_i x^i} + \sum^d_{i=0}{\beta b_i x^i}
\\
&= \sum^d_{i=0}{(\alpha a_i + \beta b_i) x^i}
\end{aligned}
$$

Como $(\alpha a_i + \beta b_i) \in \mathbb{R}$, temos que $\alpha p + \beta q \in \mathcal{P}_d$ e, por tanto, o espaço é fechado para combinação linear, sendo um subespaço vetorial de $\mathcal{F}$.

Parte 02: Mostrar que $L_X$ restrita à $\mathcal{P}_d$ é uma transformação linear

Para mostrar que é uma transformação linear basta mostrar que, se $f, g \in \mathcal{P}_d, \alpha, \beta \in \mathbb{R}$, então $L_X(\alpha f + \beta g) = \alpha L_X(f) + \beta L_X(g)$ 

$$ 
\begin{aligned}
L_X(\alpha f + \beta g) &= ((\alpha f + \beta g)(x_1),..., (\alpha f + \beta g)(x_m)) \\
&= (\alpha f(x_1),...,\alpha f(x_m)) + (\beta g(x_1),...,\beta g(x_m)) \\
&= \alpha (f(x_1),...,f(x_m)) + \beta (g(x_1),...,g(x_m)) \\
&= \alpha L_X(f) + \beta L_X(g)
\end{aligned}
$$

### Item 1.c) 
**Uma base diferente**. A base canônica de $\mathcal{P}_d$ é dada por $\set{1,x,x^2,...,x^d}$. Suponha que $m = \#X = d + 1$, e considere a família de polinômios $\set{p_1,p_2,...,p_{d+1}}$, onde $p_i$ é o polinômio de grau $d$ que satisfaz $p_i(x_j ) = δij$ para todos os $x_j ∈ X$. Mostre que eles são
linearmente independentes, e conclua que $\set{p_1,p_2,...,p_{d+1}}$ também é uma base de $\mathcal{P}_d$.


Os polinômios $p_i$ acima são conhecidos como polinômios interpoladores de Lagrange associados aos pontos de $X$.

Por absurdo:

Suponha que existam $\alpha_i \in \mathbb{R}$ tais que $\exists i$ $|$ $\alpha_i \neq 0$ e que $\sum^{d+1}_{i=1}{\alpha_i p_i} = 0$, ou seja, $\forall x \in \mathbb{R}, \sum^{d+1}_{i=1}\alpha_i p_i(x) = 0$. Entretanto, tome $x_1 \in X$, para $x_1, \sum^{d+1}_{i=1}\alpha_i p_i(x) = \alpha_1 p_1(x_1) = \alpha_1 \neq 0$. Com isso temos que $\sum^{d+1}_{i=1}\alpha_i p_i(x) = \sum^{d+1}_{i=2}\alpha_i p_i(x)$, repetindo esse processo obtemos que $ \alpha_i = 0, \forall i$, ou seja, o conjunto $\set{p_1,p_2,...,p_{d+1}}$ é $LI$, e dado que $\mathcal{P}_d$ tem dimensão $d+1$, forma uma base.

### Item 1.d) 
Use os polinômios interpoladores de Lagrange do exercício anterior para mostrar que $L_X$ é de posto completo (ou seja, $rank(L_X ) = min\set{m,d + 1})$. Separe os casos $m < d + 1$ e $m > d + 1$. O caso $m = d + 1$ é similar a qual deles?

Para $X = \set{x_1,...,x_m}$ e $L_X: \mathcal{P}_d \rightarrow \mathbb{R}^m$, onde $dim(\mathbb{R}^m) = m$ e $dim(\mathcal{P}_d) = d+1$, temos dois casos:

Caso 01: $m \le d+1$ e queremos mostrar que $L_X$ é sobrejetiva

Dado que o número total de pontos é menor que $d+1$, podemos escolher outros $d+1-m$ pontos distintos quisqueres e, com eles, montar uma base de polinômios interpoladores de Lagrange para $\mathcal{P}_d$, de modo que, qualquer $p \in \mathcal{P}_d$ possa ser escrito da forma $p = \sum^m_{i=1}{\alpha_i p_i} + \sum^{d+1}_{i=m+1}{\alpha_i p_i}$, onde $\alpha_i \in \mathbb{R}$ para toto $i$ e os $p_i$ são referentes aos $m$ pontos originais para $i=1,...,m$ e aos novs pontos para $i=m+1,...,d+1$.

Com isso, temos que, dada a linearidade de $L_X$, basta mapear os elementos da base de $\mathcal{P}_d$ para a base de $\mathbb{R}^m$. Esse mapeamento é trivial, pois $L_X(p) = (p(x_1),...,p(x_m))$ e $p_i(x_j) = \delta_{ij}$, onde $\delta_{ij} = 1$, se $i=j$ e $\delta_{ij} = 0$, se $i\neq j$, de modo que $L_X(p_i) = e_i$ para $i \le m$ e $L_X(p_i) = 0$ para $i > m$. 

Portanto, como os $m$ vetores $e_i$ são $LI$, temos uma base para $\mathbb{R}^m$ e $L_X$ é sobrejetiva.

Caso 02: $m \ge d+1$ e queremos mostrar que $L_X$ é injetiva

Similarmente ao caso 01, tomemos como base de $\mathcal{P}_d$ os polinômios interpoladores de Lagrange para os primeiros $d+1$ pontos de $X$. Logo, qualquer $p \in \mathcal{P}_d$ possa ser escrito da forma $p = \sum^m_{i=1}{\alpha_i p_i}$ onde $\alpha_i \in \mathbb{R}$ para todo $i$ e os $p_i$ são os polinômios da base referentes aos $d+1$ primeiros pontos de $X$.

Com isso, temos que $L_X(p) = (p(x_1),...,p(x_{d+1}), p(x_{d+1}),...,p(x_m))$. Para os polinômios $p_i$ da base, temos que, das $d+1$ primeiras entradas de $L_X(p_i)$, a $i$-ésima será $1$, enquanto as outras serão $0$. Logo, aplicar $L_X$ nos elementos da base de $\mathcal{P}_d$ leva a um conjunto de $d+1$ vetores $LI$, de modo que $dim(Im(L_X))=d+1$, e, portanto, $L_X$ tenha posto completo.

No caso em que $m=d+1$, caímos tanto no caso 01 quanto no 02, e, especialmente nesse caso, não é necessário adicionar nenhum ponto para completar a base do polinômios interpoladores e a saída mapeia toda a base canônica do $\mathbb{R}^m$, de modo de a transformação seja bijetora, ou seja, $rk(L_X) = m = d + 1$.

### Item 1.e) 
Conclua que uma matriz de Vandermonde é de posto completo, inversível quando $m = d + 1$. (sem precisar calcular determinantes!)

Seja $V$ a matriz de Vandermonde, queremos provar que $$\operatorname{rank}(V) = \dim(\operatorname{Im}(L_X)).$$

Parte 01: Imagem de $L_X$:

Todo $p \in \mathcal{P}_d$ pode ser escrito na base canônica como $p(x) = a_0 + a_1 x + a_2 x^2 + \cdots + a_d x^d$, de modo que a saída de $L_X$ é simplesmente o vetor das avaliações de $p$ nos pontos de $X$:
$$L_X(p) = \begin{bmatrix} p(x_1) \\ p(x_2) \\ \vdots \\ p(x_m) \end{bmatrix}.$$


Parte 02: Imagem de  $V$:

Seja $a = (a_0, a_1, \dots, a_d)^T$ o vetor de coeficientes de $p$. Multiplicando $V$ por $a$:
$$Va = \begin{bmatrix}
1 & x_1 & x_1^2 & \cdots & x_1^d \\
1 & x_2 & x_2^2 & \cdots & x_2^d \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_m & x_m^2 & \cdots & x_m^d
\end{bmatrix}
\begin{bmatrix} a_0 \\ a_1 \\ a_2 \\ \vdots \\ a_d \end{bmatrix}.$$


Fazendo a multiplicação e avaliando o produto na $i$-ésima linha temos:
$$1 \cdot a_0 + x_i \cdot a_1 + x_i^2 \cdot a_2 + \cdots + x_i^d \cdot a_d = p(x_i).$$
Essa expressão é rigorosamente a definição de avaliar $p$ no ponto $x_i$. Fazendo isso para todas as $m$ linhas, chegamos à igualdade fundamental:
$$Va = \begin{bmatrix} p(x_1) \\ p(x_2) \\ \vdots \\ p(x_m) \end{bmatrix} = L_X(p).$$
Portanto, o espaço coluna de $V$ coincide com a imagem de $L_X$, e em particular
$$\operatorname{rank}(V) = \dim(\operatorname{Im}(L_X)) = d + 1.$$

p_1(x) = \frac{(x-x_2)(x-x_3)...(x-x_{m-1})(x-x_m)}{(x_1-x_2)(x_1-x_3)...(x_1-x_{m-1})(x_1-x_m)}=\prod^m_{i=2}{\frac{(x-x_i)}{(x_1-x_i)}}

Logo, avaliar $p_1$ em $x$ exigiria apenas as $m-1$ subtrações e $m-2$ produtos do numerador, mais uma divisão final. Totalizando $2m-2$ operações aritméticas.

## **Questão 2**: Regressão polinomial
Suponha que $m ≥ d + 1$. Defina a função $R_{X,d} : \mathbb{R}^m → \mathcal{P}_d$ que associa a cada vetor $b ∈ \mathbb{R}^m$ o polinômio $p ∈ \mathcal{P}_d$ que melhor se ajusta aos valores $b_i$ (nos pontos $x_i ∈ X$) no sentido de mínimos quadrados, ou seja, $arg min_{p∈\mathcal{P}d}\|L_X (p) −b\|^2_2$.

### Item 2.a) 
Mostre que $R_{X,d}$ é uma inversa à esquerda de $L_X$ , ou seja, $R_{X,d} \circ L_X = Id_{\mathcal{P}_d}$.

A transformação $L_X$ associa um polinômio $p\in \mathcal{P}_d$ a um vetor $v\in\mathbb{R}^m$ avaliando-o em $m$ pontos distintos. No caso em que estamos analisando ($m \ge d+1$), essa transformação é injetiva (vide [item 1.d](#item-1d)), ou seja, $\forall v \in \mathbb{R}^m, \exists !\ p \in \mathcal{P}_d\ |\ L_X(p) = v$.

Logo, para um vetor $b=L_X(q)\in\mathbb{R}^m$, onde $q \in \mathcal{P}_d$, temos que $R_{X,d}(b) = argmin_{p\in\mathcal{P}_d}\|L_X(p) - b\|^2_2 = argmin_{p\in\mathcal{P}_d}\|L_X(p) - L_X(q)\|^2_2$ e, pela injetividade da $L_X$, $\|L_X(p) - L_X(q)\|^2_2 = 0 \iff p=q$. Ou seja, $R_{X,d}(b) = R_{X,d}(L_X(q)) = q$ para qualquer $q \in \mathcal{P}_d$. Portanto, podemos concluir que $R_{X,d} \circ L_X = Id_{\mathcal{P}_d}$.

Além disso, note que a transformação $R_{X,d}$ faz o processo oposto da $L_X$, levando um vetor $b\in\mathbb{R}^m$ a um polinômio $p\in \mathcal{P}_d$, especificamente, esse processo conssiste em minimização de resíduos quadrados na imagem de $L_X$, um subespaço linear do $\mathbb{R}^m$. Em especial, se lembrarmos do isomorfismo $\mathcal{P}_d \cong \mathbb{R}^{d+1}$ e tratarmos o polinômio pelos seus coeficientes $c_i$ na base canônica, de forma que possamos representá-lo por um vetor $c\in\mathbb{R}^{d+1}$, temos que a imagem de $L_X$ é igual à imagem da matriz de Vandermonde V, conforme visto no [item 1.e](#item-1e). Logo, temos um problema muito mais simples de se trabalhar em que $R'_{X,d}(b) = argmin_{c\in\mathbb{R}^{d+1}}\|Vc - b\|^2_2$. 

Nesse novo problema, já sabemos trabalhar, pois para a minimização do resíduo quadrado temos que a solução $\hat{c}$ do problema será tal que $(V\hat{c} - b) \perp Vc, \forall c\in\mathbb{R}^{d+1}$, logo:
$$
\begin{aligned}
& (V\hat{c} - b) \perp Vc, &\forall c\in\mathbb{R}^{d+1}
\\
\iff & (Vc)^*(V\hat{c} - b)=0, &\forall c\in\mathbb{R}^{d+1}
\\
\iff & V^*(V\hat{c} - b)=0
\\
\iff & V^*V\hat{c} = V^*b
\\
\iff & \hat{c} = (V^*V)^{-1}V^*b
\\
\iff & R'_{X,d} = (V^*V)^{-1}V^* = V^+
\end{aligned}
$$
Ou seja, $R_{X,d}$ atua como uma pseudo-inversa, o que indica que atuaria como inversa à direita de $L_X$ no caso onde $m \le d+1$ e $L_X$ é sobrejetiva ao invés de injetiva.

### Item 2.b) 
Se convença que $R_{X,d}$ é uma transformação linear.

Para que $R_{X,d}$ seja uma transformação linear queremos que, dados $u,v\in\mathbb{R}^m$ e $\alpha,\beta\in\mathbb{R}$, vale a seguinte igualdade:
$$
R_{X,d}(\alpha u + \beta v) = \alpha R_{X,d}(u)+\beta R_{X,d}(v)
$$
Para nos convencer disso, considere a representação de um polinômio $p\in\mathcal{P}_d$ pelos seus coeficientes na base canônica, ou seja, representando $p$ pelo vetor $c\in\mathbb{R}^{d+1}$, e podemos proceder de forma bastante simples:
$$
\begin{aligned}
R_{X,d}(\alpha u + \beta v) &= argmin_{p\in\mathcal{P}_d}\|L_X(p) - (\alpha u + \beta v)\|^2_2
\\
&= argmin_{c\in\mathbb{R}^{d+1}}\|Vc - (\alpha u + \beta v)\|^2_2
\\
&= V^+(\alpha u + \beta v)
\\
&= \alpha V^+u + \beta V^+v
\\
&= \alpha\cdot argmin_{c\in\mathbb{R}^{d+1}}\|Vc - u\|^2_2 + \beta\cdot argmin_{c\in\mathbb{R}^{d+1}}\|Vc - v\|^2_2
\\
&= \alpha R_{X,d}(u) + \beta R_{X,d}(v)
\end{aligned}
$$ 

### Item 2.c) 
Como calcular a matriz de $R_{X,d}$ na base canônica de $\mathcal{P}_d$?

Conforme discutido no [item 2.a](#item-2a), no caso onde $m\ge d+1$, temos que, na base canônica de $\mathcal{P}_d$, a transformação $R_{X,d}$ atua como pseudo-inversa à esquerda de $L_X$, assumindo a matriz $R_{X,d} = (V^*V)^{-1}V^*$, onde $V$ é a matriz de Vandermonde nos pontos de $X$.

Obs.1: no caso onde $m\le d+1$, temos que $R_{X,d}$ atua como pseudo inversa à direita de $L_X$ e, na base canônica de $\mathcal{P}_d$, $R_{X,d}=V^*(VV^*)^{-1}$

Obs.2: note que, no caso onde $m=d+1$, $V$ e $V^*$ tem inversa e, portanto, $R_{X,d} = (V^*V)^{-1}V^* = V^{-1}(V^*)^{-1}V^* = V^{-1}I = V^{-1}$ de modo que $R_{X,d}$ seja a própria inversa de $L_X$.

### Item 2.d) 
**Condicionamento**. Escolha alguns valores de $m$ e $d$ e calcule os números de condicionamento das matrizes $L_X$ e $R_{X,d}$ na base canônica de $\mathcal{P}_d$. Como o número de condicionamento das matrizes $L_X$ e $R_{X,d}$ se comporta conforme $m$ e/ou $d$ aumentam?

Conforme discutido anteriormente, a matriz da transformação $R_{X,d}$ assume a forma $R_{X,d} = (V^*V)^{-1}V^* = V^+$. Mudando um pouco a notação da matriz de Vandermonde para $\mathcal{V}$, podemos considerar sua decomposição SVD, $\mathcal{V} = U\Sigma V^*$. Com isso, podemos encontrar uma forma de calcular o condicionamento teórico $k$ dessas transformações.

Para a transformação $L_X$, representada pela matriz de Vandermonde $\mathcal{V}$, temos, conforme visto em sala, que $k(\mathcal{V})=\frac{|\sigma_1|}{|\sigma_{d+1|}}$. Note que $\sigma_{d+1}\neq 0$ pois $\mathcal{V}$ tem posto completo.

Para a transformação $R_{X,d}$, por outro lado, precisamos primeiro descobrir como calcular sua matriz: 

- Se escolhermos calcular via a resolução do sistema $(\mathcal{V}^*\mathcal{V})R_{X,d} = \mathcal{V}$, temos que considerar o condicionamento $k(\mathcal{V}^*\mathcal{V})$ ao invés de $k(\mathcal{V})$, o que será um problema, pois:
$$
\begin{aligned}
\mathcal{V}^*\mathcal{V} &= (U\Sigma V^*)^*(U\Sigma V^*) = (V\Sigma ^*U^*) U\Sigma V^* = V \hat{\Sigma}^2 V^*
\\
\Rightarrow k(\mathcal{V}^*\mathcal{V}) &= \frac{\sigma_1^2}{\sigma_{d+1}^2} = (\frac{\sigma_1}{\sigma_{d+1}})^2 = k(\mathcal{V})^2
\end{aligned}
$$

- Por outro lado, considerando a decomposição SVD de $\mathcal{V}$, podemos calcular a matriz da transformação $R_{X,d}$ diretamente:
$$
\begin{aligned}
R_{X,d} = V^+ = (\mathcal{V}^*\mathcal{V})^{-1}\mathcal{V}^* &= [(U\Sigma V^*)^*(U\Sigma V^*)]^{-1}(U\Sigma V^*)^*
\\
&= (V\Sigma^*U^*U\Sigma V^*)^{-1}V\Sigma^*U^*
\\
&= (V(\Sigma^*\Sigma)V^*)^{-1}V\Sigma^*U^*
\\
&= V(\Sigma^*\Sigma)^{-1}V^*V\Sigma^*U^*
\\
&= V(\Sigma^*\Sigma)^{-1}\Sigma^*U^*
\\
&= V\Sigma^+U^*
\end{aligned}
$$
Note que $\Sigma$ é uma matriz $m\times d+1$ com os valores singulares $\sigma_i$ de $\mathcal{V}$ na diagonal, logo, $\Sigma^+$ terá os valores $\frac{1}{\sigma_i}$ na diagonal. Desse modo, podemos calcular facilmente o condicionamento da transformação $R_{X,d}$, com base na decomposição SVD de $\mathcal{V}^+$:
$$
\begin{aligned}
k(R_{X,d}) = k({\mathcal{V}^+}) =  k(V\Sigma^+U^*) = \frac{|\frac{1}{\sigma_{d+1}}|}{|\frac{1}{\sigma_{1}}|} = \frac{|\sigma_1|}{|\sigma_{d+1|}} = k(\mathcal{V})
\end{aligned}
$$

Dado que $k(\mathcal{V}^+) = k(\mathcal(V)) = |\frac{\sigma_1}{\sigma_{d+1}}|$, para analizar o condicionamento dessas matrizes conforme $m$ e $d$ variam, precisamos avaliar como $\sigma_1$ e $\sigma_{d+1}$ se comportam. Para isso, faremos inicialmente uma simplificação do problema, limitando os pontos de $X$ a um ```linspace(-1, 1, m)```. 

Primeiramente, note que pela construção da matriz de Vandermonde os valores de uma coluna $v_i$ diminuem conforme $i$ aumenta, mais especificamente, na coluna $v_{d+1}$, o primeiro e o último valor terão módulo $1$, enquanto os outros terão o menor módulo de suas respectivas linhas e tendendo a $0$, enquanto a primeira coluna (de $1$) terá os maiores. Logo, a direção dos vetores que mais será distorcida (relativo à $\sigma_1$) será aquela relacionada à primeira entrada do vetor do domínio, enquanto a direção que será menos distorcida (relativo a $\sigma_{d+1}$) será aquela relacionada a última entrada do vetor do domínio. Com isso, temos que, para os vetores $e_1, e_{d+1} \in \mathbb{R}^{d+1}$, o "esticamento" pela transformação $\mathcal{V}$ será dado da seguinte forma:

$$
\begin{aligned}
|\sigma_1| \ge\|\mathcal{V}e_1\|_2 &= \|1 \cdot v_1\| = \sqrt{\sum^m_{k=1}{x_k^{2\cdot 0}}} = \sqrt{m} \ge \|\mathcal{V}e_k\|_2, \forall k 
\\
|\sigma_{d+1}| \le|\mathcal{V}e_{d+1}\|_2 &= \|1 \cdot v_{d+1}\| = \sqrt{\sum^m_{k=1}{x_k^{2\cdot d}}} \xrightarrow{d\rightarrow \infty} \sqrt{2}
\end{aligned}
$$

Ou seja, dado um valor de $m$, na direção de maior aumento, a distorção $\sigma_1 = \sqrt{m}$, enquanto na direção de menor aumento, a distorção $\sigma_{d+1}$ é tal que conforme $d$ cresce, seu valor diminui, pois os valores de $|x_i|^d\xrightarrow{d\rightarrow \infty}0, \forall i \not\in \{1,m\}$. Logo, o condicionamento, dado pela fração $|\frac{\sigma_1}{\sigma_{d+1}}|$ aumenta.

Por outro lado, veremos que com $m\rightarrow\infty$, não há um impacto no condicionamento de $\mathcal{V}$. Primeiramente, note que, como $k(\mathcal{V})=|\frac{\sigma_1}{\sigma_{d+1}}|$ é por uma fração, podemos dividir o numerador e o denominador por $\frac{1}{\sqrt{m}}$ e que $|\sigma_1|\ge \|\mathcal{V}e_k\|_2 \ge |\sigma_{d+1}|$, de modo que possamos analisar $ \frac{\|\mathcal{V}e_k\|_2}{\sqrt{m}} $:

$$
\frac{\|\mathcal{V}e_k\|_2}{\sqrt{m}} = \sqrt{\frac{1}{\sqrt{m}}\cdot\sum^m_{i=1}{(x_i)^{2k-1}}}\xrightarrow{m\rightarrow\infty}\sqrt{\int^1_{-1}{x^{2k-2}dx}}
$$

E especificamente:

$$
\begin{aligned}
\frac{|\sigma_1|}{\sqrt{m}} \ge \frac{\|\mathcal{V}e_1\|_2}{\sqrt{m}} &\xrightarrow{m\rightarrow\infty} \sqrt{\int^1_{-1}{x^{0}dx}} = \sqrt{2}
\\
\frac{|\sigma_{d+1}|}{\sqrt{m}} \le \frac{\|\mathcal{V}e_{d+1}\|_2}{\sqrt{m}} &\xrightarrow{m\rightarrow\infty} \sqrt{\int^1_{-1}{x^{2d}dx}} = \sqrt{\frac{1^{2d+1}-(-1)^{2d+1}}{2d+1}}
\end{aligned}
$$

Logo, temos que, $k(\mathcal{V})=|\frac{\sigma_1}{\sigma_{d+1}}|=|\frac{\frac{\sigma_1}{\sqrt{m}}}{\frac{\sigma_{d+1}}{\sqrt{m}}}|$, e que, conforme o valor de $m$ cresce, tendemos a uma integral, o que não depende do prório $m$, essa variável não deve impactar diretamente no condicionamento, como poderemos ver na análise dos gráficos a seguir:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def cond_LX_RXd(m, d):
    V = np.vander(np.linspace(-1, 1, num=m), d+1, increasing=True)
    return {'m':m, 'd':d, 'L_X':np.linalg.cond(V), 'R_Xd':np.linalg.cond(np.linalg.pinv(V))}

cond = []
for d in range(100):
    for m in range(d+1, 500):
        cond.append(cond_LX_RXd(m, d))

dfc = pd.DataFrame(cond)
````

````python
from matplotlib.colors import LogNorm

display(dfc.head())

lx_c_df = dfc.pivot(index='d', columns='m', values='L_X')
sns.heatmap(lx_c_df, norm=LogNorm())
plt.title('Variação do condicionamento conforme m e d+1')
plt.ylabel('d')
plt.xlabel('m')
````
