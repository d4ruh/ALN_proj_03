## **Questão 1**: Polinômios interpoladores
Nesta questão, iremos trabalhar com o espaço de funções reais $\mathcal{F}$, e também com o espaço dos polinômios de grau até $d$, denotado por $\mathcal{P}_d$.

a) O **funcional de avaliação**. Seja $X = \set{x_1,...,x_m}$ um conjunto de pontos distintos em $\mathbb{R}$. Defina o funcional de avaliação $L_X : \mathcal{F} → \mathbb{R}^m$ por $L_X (f) = (f(x_1),...,f(x_m))$. Mostre que $L_X$ é uma transformação linear.

Para mostrar que $L_X$ é uma transformação linear basta mostrar que $L_X(\alpha f + \beta g) = \alpha L_X(f) + \beta L_X(g), \forall \alpha, \beta \in \mathbb{R}, f, g \in \mathcal{F}$ 

$$ 
\begin{aligned}
L_X(\alpha f + \beta g) &= ((\alpha f + \beta g)(x_1),..., (\alpha f + \beta g)(x_m)) \\
&= (\alpha f(x_1),...,\alpha f(x_m)) + (\beta g(x_1),...,\beta g(x_m)) \\
&= \alpha (f(x_1),...,f(x_m)) + \beta (g(x_1),...,g(x_m)) \\
&= \alpha L_X(f) + \beta L_X(g)
\end{aligned}
$$

b) Podemos ver o espaço dos polinômios $\mathcal{P}_d$ como um subconjunto de $\mathcal{F}$. Mostre que de fato é um subespaço linear, e conclua que $L_X$ restrita a $\mathcal{P}_d$ é uma transformação linear. (Dica: não é necessário usar uma base de $\mathcal{P}_d$!)

Parte 01: Mostrar que $\mathcal{P}_d$ é um subespaço linear

Sejam $p, q \in \mathcal{P}_d$, logo ambos tem grau menor ou igual à $d$ e, para $a_i, b_i \in \mathbb{R}$, podemos escreve-los da seguinte forma:
$$
p(x) = \sum^d_{i=0}{a_i x^i}
\\
q(x) = \sum^d_{i=0}{b_i x^i}
$$

Agora, dados $\alpha, \beta \in \mathbb{R}$, temos que:
$$
\begin{aligned}
\alpha p(x) + \beta q(x) &= \alpha \sum^d_{i=0}{a_i x^i} + \beta \sum^d_{i=0}{b_i x^i} 
\\
&= \sum^d_{i=0}{\alpha a_i x^i} + \sum^d_{i=0}{\beta b_i x^i}
\\
&= \sum^d_{i=0}{(\alpha a_i + \beta b_i) x^i}
\end{aligned}
$$

Como $(\alpha a_i + \beta b_i) \in \mathbb{R}$, temos que $\alpha p + \beta q \in \mathcal{P}_d$ e, por tanto, o espaço é fechado para combinação linear, sendo um subespaço vetorial de $\mathcal{F}$.

Parte 02: Mostrar que $L_X$ restrita à $\mathcal{P}_d$ é uma transformação linear

Para mostrar que é uma transformação linear basta mostrar que, se $f, g \in \mathcal{P}_d, \alpha, \beta \in \mathbb{R}$, então $L_X(\alpha f + \beta g) = \alpha L_X(f) + \beta L_X(g)$ 

$$ 
\begin{aligned}
L_X(\alpha f + \beta g) &= ((\alpha f + \beta g)(x_1),..., (\alpha f + \beta g)(x_m)) \\
&= (\alpha f(x_1),...,\alpha f(x_m)) + (\beta g(x_1),...,\beta g(x_m)) \\
&= \alpha (f(x_1),...,f(x_m)) + \beta (g(x_1),...,g(x_m)) \\
&= \alpha L_X(f) + \beta L_X(g)
\end{aligned}
$$

c) **Uma base diferente**. A base canônica de $\mathcal{P}_d$ é dada por $\set{1,x,x^2,...,x^d}$. Suponha que $m = \#X = d + 1$, e considere a família de polinômios $\set{p_1,p_2,...,p_{d+1}}$, onde $p_i$ é o polinômio de grau $d$ que satisfaz $p_i(x_j ) = δij$ para todos os $x_j ∈ X$. Mostre que eles são
linearmente independentes, e conclua que $\set{p_1,p_2,...,p_{d+1}}$ também é uma base de $\mathcal{P}_d$.


Os polinômios $p_i$ acima são conhecidos como polinômios interpoladores de Lagrange associados aos pontos de $X$.

Por absurdo:

Suponha que existam $\alpha_i \in \mathbb{R}$ tais que $\exists i$ $|$ $\alpha_i \neq 0$ e que $\sum^{d+1}_{i=1}{\alpha_i p_i} = 0$, ou seja, $\forall x \in \mathbb{R}, \sum^{d+1}_{i=1}\alpha_i p_i(x) = 0$. Entretanto, tome $x_1 \in X$, para $x_1, \sum^{d+1}_{i=1}\alpha_i p_i(x) = \alpha_1 p_1(x_1) = \alpha_1 \neq 0$. Com isso temos que $\sum^{d+1}_{i=1}\alpha_i p_i(x) = \sum^{d+1}_{i=2}\alpha_i p_i(x)$, repetindo esse processo obtemos que $ \alpha_i = 0, \forall i$, ou seja, o conjunto $\set{p_1,p_2,...,p_{d+1}}$ é $LI$, e dado que $\mathcal{P}_d$ tem dimensão $d+1$, forma uma base.

d) Use os polinômios interpoladores de Lagrange do exercício anterior para mostrar que $L_X$ é de posto completo (ou seja, $rank(L_X ) = min\set{m,d + 1})$. Separe os casos $m < d + 1$ e $m > d + 1$. O caso $m = d + 1$ é similar a qual deles?

Para $X = \set{x_1,...,x_m}$ e $L_X: \mathcal{P}_d \rightarrow \mathbb{R}^m$, onde $dim(\mathbb{R}^m) = m$ e $dim(\mathcal{P}_d) = d+1$, temos dois casos:

Caso 01: $m \le d+1$ e queremos mostrar que $L_X$ é sobrejetiva

Dado que o número total de pontos é menor que $d+1$, podemos escolher outros $d+1-m$ pontos distintos quisqueres e, com eles, montar uma base de polinômios interpoladores de Lagrange para $\mathcal{P}_d$, de modo que, qualquer $p \in \mathcal{P}_d$ possa ser escrito da forma $p = \sum^m_{i=1}{\alpha_i p_i} + \sum^{d+1}_{i=m+1}{\alpha_i p_i}$, onde $\alpha_i \in \mathbb{R}$ para toto $i$ e os $p_i$ são referentes aos $m$ pontos originais para $i=1,...,m$ e aos novs pontos para $i=m+1,...,d+1$.

Com isso, temos que, dada a linearidade de $L_X$, basta mapear os elementos da base de $\mathcal{P}_d$ para a base de $\mathbb{R}^m$. Esse mapeamento é trivial, pois $L_X(p) = (p(x_1),...,p(x_m))$ e $p_i(x_j) = \delta_{ij}$, onde $\delta_{ij} = 1$, se $i=j$ e $\delta_{ij} = 0$, se $i\neq j$, de modo que $L_X(p_i) = e_i$ para $i \le m$ e $L_X(p_i) = 0$ para $i > m$. 

Portanto, como os $m$ vetores $e_i$ são $LI$, temos uma base para $\mathbb{R}^m$ e $L_X$ é sobrejetiva.

Caso 02: $m \ge d+1$ e queremos mostrar que $L_X$ é injetiva

Similarmente ao caso 01, tomemos como base de $\mathcal{P}_d$ os polinômios interpoladores de Lagrange para os primeiros $d+1$ pontos de $X$. Logo, qualquer $p \in \mathcal{P}_d$ possa ser escrito da forma $p = \sum^m_{i=1}{\alpha_i p_i}$ onde $\alpha_i \in \mathbb{R}$ para toto $i$ e os $p_i$ são os polinômios da base referentes aos $d+1$ primeiros pontos de $X$.

Com isso, temos que $L_X(p) = (p(x_1),...,p(x_{d+1}), p(x_{d+1}),...,p(x_m))$. Para os polinômios $p_i$ da base, temos que, das $d+1$ primeiras entradas de $L_X(p_i)$, a $i-ésima$ será $1$, enquanto as outras serão $0$. Logo, aplicar $L_X$ nos elementos da base de $\mathcal{P}_d$ leva a um conjunto de $d+1$ vetores $LI$, de modo que $dim(Im(L_X))=d+1$, e, portanto, $L_X$ tenha posto completo.
